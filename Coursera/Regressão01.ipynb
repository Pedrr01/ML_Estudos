Irei criar dois arrays numpy eles: (arrays multidimensionais que são matrizes)
variável X -> entrada (tamanho)
variável Y -> objetivo do modelo (preço)
cada uma com dois valores dentro

```python
x = np.array([1.0, 2.0])
y = np.array([300.0, 500.0])
print(f"Eixo X = {x}")
print(f"Eixo Y = {y}")
```

Para mostrar a quantidade de elementos da minha variável:
Uso o função .shape ou posso usar len(nome)

```python
#Uso o .shape ou posso usar o len(nome)
print(f"Quantidade de elementos da variável X: {x.shape}")
#Para exibir a quantidade de elementos de um array específico .shape[posição]
m = x.shape[0]
print(f"Quantidades de elemento na primeria dimensão do array: {m}")
```

```python
i = 0 #Variável iniciada em 0 que será usada para acessar os valores dos arrays

x_i = x[i]
y_i = y[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

Criando o Gráfico:
a função scatter faz a criação de um gráfico de ponto em um plano cartesiano,

```python
plt.scatter(x, y, marker='x', color="red")

#Rótulos para o gráfico
plt.title("Preço Das Residências")
plt.ylabel('Preço')
plt.xlabel('Tamanho')

#A função .show é usada para exibir o gráfico
plt.show()
```

Montando a Função de Regressão Linear:
inicio as variáveis com 100 

```python
a = 100 #Coef. Angular
b = 100 #Coef. Linear
print(f"a: {a}")
print(f"b: {b}")
```

Função com o calculo:
recebe o array X

```python
def regressão(x, a, b):
    m = x.shape[0]
    função = np.zeros(m)
    for i in range(m):
        função[i] = a * x[i] + b
        
    return função
```

```python
mapeamento = regressão(x, a, b,)

plt.plot(x, mapeamento, c='b',label='Previsão')
plt.scatter(x, y, marker='x', c='r',label='Valores Atuais')

plt.title("Preços das Residências")
plt.ylabel('Preço')
plt.xlabel('Tamanho')

plt.legend()
plt.show()
```
