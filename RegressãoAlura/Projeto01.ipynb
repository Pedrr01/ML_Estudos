Importação das Bibliotecas:

```python
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
```

Selecionando o local do arquivo:
(e definindo a separação a cada ";")

```python
caminhoArquivo = r"D:\PedroPC\Documents\EstudosML\RegressãoAlura\Dados\Consumo_cerveja.csv"
dados = pd.read_csv(caminhoArquivo, sep=";")
```

```python
dados_numericos = dados.drop('data', axis=1)
matriz_correlacao = dados_numericos.corr()
matriz_correlacao = matriz_correlacao.round(4)
print(matriz_correlacao)
```

Análises Graficas:

```python
fig, ax = plt.subplots(figsize=(13, 5))

ax.set_title("Consumo de Cerveja", fontsize=18, fontweight='bold', color='black')
ax.set_ylabel("Litros", fontsize=16, color='black')
ax.set_xlabel("Dias do Ano", fontsize=16, color='black')

ax = dados["consumo"].plot(color='orange')
```

Gráfico Boxplot ou Diagrama da Caixa:
(Apenas uma variável)

```python
import seaborn as sns

ax = sns.boxenplot(data=dados["consumo"], orient="v", width=0.2, color="orange")
ax.figure.set_size_inches(13, 5)
ax.set_title("Consumo de Cerveja", fontsize=18)
ax.set_ylabel("Litros", fontsize=16)
```

Gráfico Boxplot ou Diagrama da Caixa:
(Com duas variáveis onde -> 0 = meio de semana / 1 = fim de semana)

```python
sns.set_palette("Accent_r")
sns.set_style("darkgrid")

ax = sns.boxenplot(y="consumo", x="fds", data=dados, orient="v", width=0.5)
ax.figure.set_size_inches(13, 5)
ax.set_title("Consumo de Cerveja", fontsize=20)
ax.set_ylabel("Litros", fontsize=16)
ax.set_xlabel("Final de Semana", fontsize=16)

```

Fazendo histograma (Distribuição de frequências):

```python
ax = sns.displot(dados["consumo"])
ax.figure.set_size_inches(13, 5)
ax.set_axis_labels("Consumo de Cerveja (LITROS)", "Distribuição de Frequências", fontsize=16)
```
